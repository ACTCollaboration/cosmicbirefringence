```python
import numpy as np, basic, prjlib, plottools as pl, binning as bins
from matplotlib.pyplot import *
```

```python
mb = bins.multipole_binning(12,spc='p2',lmin=20,lmax=3000)
ps, __ = prjlib.filename_init(doreal='True',rlmin='200',dearot='False',ver='v1',qtagext='_oL3000')
scb = prjlib.binned_cl_rlz(ps.quad.f['EB'].cl,1,100,mb)
vcb = np.std(scb,axis=0)
L, n0 = np.loadtxt(ps.quad.f['EB'].n0bs,unpack=True,usecols=(0,1))
```

```python
vb = np.sqrt(bins.binning1(n0**2/(L+.5),mb))
vb = vb/(mb.bp[1:]-mb.bp[:-1])**.5*10.
pl.plot_1dstyle(ylab=r'Error (Analytic/Simulation)',spc='p2',xmin=20,xmax=3000,ymin=0.,ymax=1.2)
plot(np.sqrt(mb.bc),vb/vcb,label=r'analytic $1\sigma$ error')
legend(loc=0)
```
