```python
import numpy as np, healpy as hp
from matplotlib.pyplot import *
# cmblensplus
import analysis as ana, cmb
# local
import prjlib
```

```python
R = cmb.Int_dust(150)/cmb.Int_dust(353)
C = 283./372. # MJy/sr <-> K_CMB conversion factor (avg)
print('total scaliing',R*C)
```

```python
__, f = prjlib.filename_init(doreal='False')
Tgal = hp.fitsfunc.read_map(f.dust,field=0,verbose=False)*R*C/2.726
Qgal = hp.fitsfunc.read_map(f.dust,field=1,verbose=False)*R*C/2.726
Ugal = hp.fitsfunc.read_map(f.dust,field=2,verbose=False)*R*C/2.726
M = hp.fitsfunc.read_map(f.amask,verbose=False)
rM = hp.fitsfunc.read_map(f.rmask,verbose=False)
W2 = np.average(rM**2*M**2)
w2 = np.average(M**2)
print(W2,w2)
```

```python
T = ana.change_coord(Tgal,coord=['G','C'])
Q = ana.change_coord(Qgal,coord=['G','C'])
U = ana.change_coord(Ugal,coord=['G','C'])
```

```python
v = 1e-5
hp.mollview(Q,min=0,max=v)
hp.mollview(U,min=0,max=v)
```
