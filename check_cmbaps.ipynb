```python
import numpy as np, healpy as hp, analysis as ana, plottools as pl, binning as bins, prjlib
from matplotlib.pyplot import *
```

```python
ps, fs = prjlib.filename_init(doreal='True',stype='lcmb',dearot='False')
mb = bins.multipole_binning(50,lmin=200,lmax=3000)
simn = 20
```

```python
sTT = prjlib.binned_cl_rlz(fs.cli,1,simn,mb,cn=1)
mTT = np.mean(sTT,axis=0)
vTT = np.std(sTT,axis=0)
oTT = prjlib.binned_cl(fs.cli[0],mb,cn=1)
pl.plot_1dstyle(ylab=r'$C_L$',xmin=100,xmax=3000)
errorbar(mb.bc,oTT,yerr=vTT,fmt='o',label='obs')
plot(mb.bc,mTT,label='sim')
legend(loc=0,frameon=False)
```

```python
sEE = prjlib.binned_cl_rlz(fs.cli,1,simn,mb,cn=2)
mEE = np.mean(sEE,axis=0)
vEE = np.std(sEE,axis=0)
oEE = prjlib.binned_cl(fs.cli[0],mb,cn=2)
pl.plot_1dstyle(ylab=r'$C_L$',xmin=100,xmax=3000)
errorbar(mb.bc,oEE,yerr=vEE,fmt='o',label='obs')
plot(mb.bc,mEE,label='sim')
legend(loc=0,frameon=False)
```

```python
for sn in [100,300,500]:
    mb = bins.multipole_binning(10,lmin=200,lmax=2048)
    sEE = prjlib.binned_cl_rlz(fs.cli,1,sn,mb,cn=2)
    sBB = prjlib.binned_cl_rlz(fs.cli,1,sn,mb,cn=3)
    sEB = prjlib.binned_cl_rlz(fs.cli,1,sn,mb,cn=6)
    amp = sEB/(sEE-sBB)
    pl.plot_corr(amp,20,2048,clab='corr. coeff.',zmin=-.5,zmax=.5)
```
