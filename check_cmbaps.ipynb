```python
import numpy as np, healpy as hp, analysis as ana, plottools as pl, binning as bins, prjlib
from matplotlib.pyplot import *
```

```python
# initial set up
mb = bins.multipole_binning(50,lmin=200,lmax=3000)
simn = 200
```

```python
# cmb angular power spectrum
for patch in ['deep56','boss']:
    p, f = prjlib.filename_init(doreal='True',stype='lcmb',dearot='False',PSA='s14&15_'+patch)
    for aps, cn in [('TT',1),('EE',2),('BB',3),('TE',4),('TB',5),('EB',6)]:
        ocb = prjlib.binned_cl(f.cli[0],mb,cn=cn)
        scb = prjlib.binned_cl_rlz(f.cli,1,simn,mb,cn=cn)
        mcb = np.mean(scb,axis=0)
        vcb = np.std(scb,axis=0)
        st = ana.PTEs(ocb,scb,comment=aps+','+patch)
        pl.plot_1dstyle(ylab=r'$C_L^{'+aps+'}$',xmin=100,xmax=3000)
        errorbar(mb.bc,ocb,yerr=vcb,fmt='o',label='obs')
        plot(mb.bc,mcb,label='sim')
        legend(loc=0,frameon=False)
        show()
        clf()
```
