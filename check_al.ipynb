```python
import numpy as np, prjlib, plottools as pl
from matplotlib.pyplot import *
```

```python
p0, f0 = prjlib.filename_init(doreal='True',stype='lcmb',rlmin='200')
p1, f1 = prjlib.filename_init(doreal='True',stype='lcmb',rlmin='300')
p2, f2 = prjlib.filename_init(doreal='True',stype='lcmb',rlmin='400')
p3, f3 = prjlib.filename_init(doreal='True',stype='lcmb',rlmin='500')
L, al0 = np.loadtxt(p0.quad.f['EB'].al,usecols=(0,1),unpack=True)[:,2:]
L, al1 = np.loadtxt(p1.quad.f['EB'].al,usecols=(0,1),unpack=True)[:,2:]
L, al2 = np.loadtxt(p2.quad.f['EB'].al,usecols=(0,1),unpack=True)[:,2:]
L, al3 = np.loadtxt(p3.quad.f['EB'].al,usecols=(0,1),unpack=True)[:,2:]
pl.plot_1dstyle(ylab=r'$N_L$',xmin=20,xmax=2048,ymin=1.,ymax=2.)
plot(L,al1/al0,label='300/200')
plot(L,al2/al0,label='400/200')
plot(L,al3/al0,label='500/200')
legend(loc=0,frameon=False)
savefig('noise_comp.png')
```
