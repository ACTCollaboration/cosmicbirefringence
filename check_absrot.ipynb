```python
import numpy as np, prjlib, binning as bins, plottools as pl
```

```python
# correlation matrix
p, f = prjlib.filename_init(doreal='True',stype='lcmb',dearot='False')
mb = bins.multipole_binning(50,lmin=200,lmax=3000)
for sn in [100,300,500]:
    mb = bins.multipole_binning(10,lmin=200,lmax=2048)
    sEE = prjlib.binned_cl_rlz(f.cli,1,sn,mb,cn=2)
    sBB = prjlib.binned_cl_rlz(f.cli,1,sn,mb,cn=3)
    sEB = prjlib.binned_cl_rlz(f.cli,1,sn,mb,cn=6)
    amp = sEB/(sEE-sBB)
    pl.plot_corr(amp,mb.bp,clab='corr. coeff.',zmin=-.5,zmax=.5)
```
