```python
import numpy as np, prjlib, binning as bins, plottools as pl
from matplotlib.pyplot import *
```

```python
p, __ = prjlib.filename_init(stype='a0p3',rlmin='200')
L, claa, xlaa, xvaa = np.loadtxt(p.quad.f['EB'].mcls,usecols=(0,1,4,8),unpack=True)[:,1:]
```

```python
pl.plot_1dstyle(xlab='$L$',ylab=r'$C_L$',xmin=20,xmax=2048,ymin=1e-9,ymax=1e-6,xylog=True)
plot(L,claa,label=r'recons x recons')
plot(L,xlaa,label=r'input x recons')
plot(L,2.*np.pi/L/(L+1.)*1e-4*0.3,label=r'Scale-invariant spectrum with $A_{CB}=0.3$')
legend(loc=0,frameon=False)
```

```python
pl.plot_1dstyle(spc='p2',xlab='$L$',ylab=r'rec x inp / inp x inp (binned spectrum)',xmin=20,xmax=2048,ymin=.99,ymax=1.01)
axhline(1,color='k',ls='--')
mb = bins.multipole_binning(10,spc='p2',lmin=20)
scb = prjlib.binned_cl_rlz(p.quad.f['EB'].cl,1,100,mb,cn=4)
fcb = prjlib.binned_claa(2048,mb)*.3
plot(np.sqrt(mb.bc),np.mean(scb,axis=0)/fcb)
```
