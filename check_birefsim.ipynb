```python
import numpy as np, prjlib, plottools as pl
from matplotlib.pyplot import *
```

```python
patch = 'deep56'
ps, fs = prjlib.filename_init(stype='lcmb',doreal='False',PSA='s14&15_'+patch)
pa, fa = prjlib.filename_init(stype='absrot',PSA='s14&15_'+patch)
l, EEs, BBs, TEs = np.loadtxt(fs.scl,usecols=(0,2,3,4),unpack=True)
l, TBi, EBi = np.loadtxt(fs.cli[1],usecols=(0,5,6),unpack=True)
l, TBa, EBa = np.loadtxt(fa.cli[1],usecols=(0,5,6),unpack=True)
```

```python
pl.plot_1dstyle(ylab=r'$C_L^{EB}$',xmin=100,xmax=3000,grid=True)
alpha = -0.08*np.pi/180.
plot(l,EBa-EBi,label=r'global rotation sim with $\alpha=-0.08$deg')
plot(l,(EEs-BBs)*2*alpha,label=r'$2\alpha (EE-BB)$ with $\alpha=-0.08$deg')
legend(loc=0)
#savefig('fig_EB.png')
```

```python
pl.plot_1dstyle(ylab=r'$C_L^{TB}$',xmin=100,xmax=3000,grid=True)
plot(l,TBa-TBi,label=r'global rotation sim with $\alpha=-0.08$deg')
plot(l,(TEs)*2*alpha,label=r'$2\alpha (EE-BB)$ with $\alpha=-0.08$deg')
legend(loc=0)
```
