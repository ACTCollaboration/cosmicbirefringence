```python
import numpy as np
from matplotlib.pyplot import *
import prjlib
```

```python
patch = 'deep56'
ps, fs = prjlib.filename_init(stype='lcmb',doreal='False',PSA='s14&15_'+patch)
pa, fa = prjlib.filename_init(stype='absrot',PSA='s14&15_'+patch)
l, EEs, BBs, EBs = np.loadtxt(fs.scl,usecols=(0,2,3,6),unpack=True)
l, EBs = np.loadtxt(fs.ocl,usecols=(0,6),unpack=True)
l, EBa = np.loadtxt(fa.ocl,usecols=(0,6),unpack=True)
```

```python
alpha = -0.08*np.pi/180.
grid(True)
xlim(100,3000)
xlabel('$L$')
ylabel(r'$C_L^{\alpha\alpha}$')
plot(l,(EEs-BBs)*2*alpha,label=r'$2\alpha (EE-BB)$ with $\alpha=0.08$deg')
plot(l,EBa-EBs,label=r'global rotation sim with $\alpha=0.08$deg')
legend(loc=0)
savefig('fig_EB.png')
```
